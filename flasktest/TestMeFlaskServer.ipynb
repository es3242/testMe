```python
from flask import Flask, request, jsonify
from flask_cors import CORS
import json
import requests
from pdfminer.high_level import extract_pages
from typing import List
from pdfminer.layout import LTTextContainer

app = Flask(__name__)
CORS(app)

# 추출된 텍스트를 저장할 리스트
dodolist = []
url = "http://localhost:5000/Blank_Fill"
headers = {
    "Content-Type": "application/json"
}

@app.route('/endpoint', methods=['POST'])
def endpoint():
    # 요청에서 JSON 데이터 가져오기
    json_data = request.get_json()
    if json_data is None:
        return '잘못된 JSON 데이터입니다', 400

    # JSON에서 'text' 파라미터 추출하기
    extracted_text = json_data.get('text')
    if extracted_text is None:
        return '필수 파라미터인 text가 누락되었습니다', 400

    # JSON에서 'page' 파라미터 추출하기
    page_num = json_data.get('page')

    # 추출된 텍스트를 dodolist에 추가하기
    dodolist.append(extracted_text)
    
    # 요청 데이터 생성
    data = {
        "Pages" : len(dodolist),
        "Text_based_PDF": dodolist
    }

    # 서버에 요청 보내기 (post로 접근)
    response = requests.post(url, headers=headers, data=json.dumps(data)).json()
    Test = response['Test']
    Solutions = response['Solutions']
            
    # 응답용 JSON 데이터 준비하기
    response_data = {
        'status': '성공',
        'page': page_num,
        'text': extracted_text,
        'Test': Test,
        'Solutions': Solutions
    }

    return jsonify(response_data)

@app.route('/dodolist', methods=['GET'])
def get_dodolist():
    data = {
        "Pages": len(dodolist),
        "Text_based_PDF": dodolist
    }

    # CORS 허용 헤더 설정
    response = jsonify(data)
    response.headers.add('Access-Control-Allow-Origin', '*')
    return response

if __name__ == '__main__':
    app.run(host='localhost', port=5000, debug=False)
```

```python
import requests
import json
from pdfminer.high_level import extract_pages
from typing import List
from pdfminer.layout import LTTextContainer
```

```python
def extract_text_pages(input_file: str) -> List[str]:
    """
    PDF 파일로부터 각 페이지마다 추출한 텍스트를 리스트로 반환.
    """
    extracted_text_pages = []
    for page_layout in extract_pages(input_file):
        page_text = ''
        for element in page_layout:
            if isinstance(element, LTTextContainer):
                page_text += element.get_text()
        if page_text:
            extracted_text_pages.append(page_text)

    return extracted_text_pages
```

<h2>단답식 test

```python
# 서버 주소와 포트 번호 설정
url = "http://localhost:5080/Blank_Fill"
```

```python
Text_based_PDF = extract_text_pages('test.pdf')
```

```python
# 요청 데이터 생성
data = {
    "Pages" : len(Text_based_PDF),
    "Text_based_PDF": Text_based_PDF
    
}
headers = {
    "Content-Type": "application/json"
}

# 서버에 요청 보내기 (post로 접근)
# response = requests.post(url, headers=headers, data=json.dumps(data)).json()
response = requests.post(url, headers=headers, data=json.dumps(data))
print(response)
```

```python
Test = response['Test']
Solutions = response['Solutions']
Questions_Amount = response['Questions Amount']
```

```python
# 응답 결과 출력
for i in range(Questions_Amount):
    print('-----------------------------------------------')
    print(f'Test{i}\n')
    print(f"Questions:{Test[i]}\n Answer:{Solutions[i]}")
    print('-----------------------------------------------')
```

<h2>객관식 test

```python
url = "http://localhost:5000/MultipleChoice"
```

```python
Questions_Amount = 1 #추출할 객관식 문제 수
```

```python
Text_based_PDF = extract_text_pages('test.pdf')
```

```python
# 요청 데이터 생성
data = {
    "Pages" : len(Text_based_PDF),
    "Text_based_PDF": Text_based_PDF,
    "Questions_Amount":Questions_Amount
    
}
headers = {
    "Content-Type": "application/json"
}

# 서버에 요청 보내기 (post로 접근)
response = requests.post(url, headers=headers, data=json.dumps(data)).json()
```

```python
Test = response['Test']
```

```python
def get_dict_from_request(Test):
    questions = []

    for item in Test:
        # 문제 파싱
        questions.append(eval(item))

    return questions
```

```python
questions = get_dict_from_request(Test)
```

```python
for i in range(Questions_Amount):
    print('-------------------------')
    print(f'객관식 {i}번:\n')
    print(f'문제: {questions[i]["question"]}\n')
    print(f'1번: {questions[i]["d1"]}\n')
    print(f'2번: {questions[i]["d2"]}\n')
    print(f'3번: {questions[i]["d3"]}\n')
    print(f'4번: {questions[i]["d4"]}\n')
    print('-------------------------')
    print(f'정답: {questions[i]["solution"]}\n')

```
